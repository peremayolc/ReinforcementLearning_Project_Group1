```python
import json
import matplotlib.pyplot as plt

# Function to load JSON data
def load_json(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

# Load data from the two files
file_len_episode = 'A2CCnnAssault_7000000ep_len_mean.json'
file_ep_reward = 'A2CCnnAssault_7000000ep_rew_mean.json'

len_episode_data = load_json(file_len_episode)
ep_reward_data = load_json(file_ep_reward)

# Extract x (timesteps) and y (values) from JSON data
def extract_xy(data):
    timesteps = [entry[1] for entry in data]  # Timesteps
    values = [entry[2] for entry in data]     # Episode values
    return timesteps, values

# Extract data for plots
len_episode_x, len_episode_y = extract_xy(len_episode_data)
ep_reward_x, ep_reward_y = extract_xy(ep_reward_data)

# Create plots
plt.figure(figsize=(12, 6))

# Plot 1: Length of Episode
plt.subplot(1, 2, 1)
plt.plot(len_episode_x, len_episode_y, label='Episode Length', color='blue')
plt.xlabel('Timesteps')
plt.ylabel('Length of Episode')
plt.title('Episode Length Over Time')
plt.grid()
plt.legend()

# Plot 2: Episode Reward
plt.subplot(1, 2, 2)
plt.plot(ep_reward_x, ep_reward_y, label='Episode Reward', color='green')
plt.xlabel('Timesteps')
plt.ylabel('Episode Reward')
plt.title('Episode Reward Over Time')
plt.grid()
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
